```
import numpy as np
import pandas as pd
```

```
import re
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```
from sklearn.feature_extraction.text import TfidfVectorizer
```

```
df = pd.read_csv('quora_duplicate_questions.tsv', delimiter='\t')
df.head(5)
```

```
df = df.drop(['id', 'qid1', 'qid2', 'is_duplicate'], axis=1)
df.head()
```

```
new_df = df['question1']
new_df = new_df.append(df['question2'])
new_df.shape
```

```
corpus = []
for question in new_df:
  question = re.sub('[^a-zA-Z]', ' ', str(question))
  question = question.lower()
  question = question.split()
  question = ' '.join(question)
  corpus.append(question)
```

```
pd.DataFrame(corpus).head(10)
```

```
tv = TfidfVectorizer()
tv = tv.fit(corpus)
X = tv.transform(corpus)
```

```
tfidf = dict(zip(tv.get_feature_names(), tv.idf_))
print(tfidf)
```
